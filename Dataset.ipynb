```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('/content/skills.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df[['Commodity Title']]
```

```python
df.head()
```

```python
# Define the word to be removed
word_to_remove = 'software'

# Remove the word from the specified column
df['Commodity Title'] = df['Commodity Title'].str.replace(word_to_remove, '', regex=False)


```

```python
df.head()
```

```python
# prompt: Using dataframe df: i want frequency  of all the commodity titles

df['Commodity Title'].value_counts() # Counts the frequency of each unique value in the 'Commodity Title' column
```

```python
# prompt: create a new column frequency to store the frequency value



frequency = df['Commodity Title'].value_counts()
df_frequency = pd.DataFrame({'Commodity Title': frequency.index, 'Frequency': frequency.values})
print(df_frequency)
```

```python
df_frequency.head()
```

```python
# prompt: store above dataframe into new csv file

df_frequency.to_csv('frequency.csv', index=False)
```
